```python
import pandas as pd

# Чтение данных
jokes_data = pd.read_csv('input.csv')
print(jokes_data.head())

# Очистка текста и приведение к нижнему регистру
jokes_data['text_clean'] = jokes_data['text'].replace(r'[^\w\s]', ' ', regex=True).replace(r'\s+', ' ', regex=True).str.lower()

# Обработка текстов с использованием spaCy
import spacy
from spacy import load
from spacy.lang.ru import Russian
!python -m spacy download ru_core_news_sm

nlp = Russian()
load_model = load("ru_core_news_sm")
lemma = []

for doc in load_model.pipe(jokes_data["text_clean"].values):
    lemma.append([n.lemma_ for n in doc])

# Добавление лемм в датафрейм
jokes_data['text_clean_lemma'] = lemma
jokes_data['text_clean_lemma_as_str'] = [' '.join(map(str, l)) for l in jokes_data['text_clean_lemma']]

# Удаление стоп-слов
!pip install nltk
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stopwords_ru = stopwords.words("russian")

jokes_data['text_clean_lemma'] = jokes_data['text_clean_lemma'].apply(lambda x: [item for item in x if item not in stopwords_ru])
jokes_data['text_clean_lemma_as_str'] = [' '.join(map(str, l)) for l in jokes_data['text_clean_lemma']]
jokes_data['text_clean_lemma_as_str']
```
